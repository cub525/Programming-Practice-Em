```python
def cmath_arrow(start, end,**kwargs):
    plt.arrow(start.real,start.imag,end.real,end.imag,**kwargs)
```

## 5.8

```python

wc = 1130*g
nt_x = math.sin(math.radians(25))*wc
nt_r = nt_x/math.cos(math.radians(31))
nt = cmath.rect(nt_r,math.radians(25+31))
z = 0+0*1j
nf = nt+ 0-wc*1j
nf_r,_ = cmath.polar(nf) 
my_fig = plt.figure()
plt.axis('equal')
plt.axis('off')
ax = plt.arrow(0,0,0,-wc,head_width = 500,label = 'Fg', color = 'r')
cmath_arrow(0+0*1j,nt,head_width = 500,label = 'Ff',color = 'b')
cmath_arrow(z,-nf,head_width = 500,label = 'Fn',color = 'y')
plt.title('FBD of Car (to scale)')
plt.xlabel('N')
plt.ylabel('N')
plt.legend()
print('Fg = {0} N at -90°, Fn = {1} N at 115°, Ft = {2} N at 56°'.format(round(wc,2),round(nf_r,2),round(nt_r,2)))
```
